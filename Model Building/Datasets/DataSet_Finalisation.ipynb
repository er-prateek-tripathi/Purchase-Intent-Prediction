```python

Weekend_days = ['Friday', 'Saturday', 'Sunday']

def get_weekday(row):
    if row['Weekend']:
        return random.choice(Weekend_days)
    else:
        weekday_name = row['Date_Time'].strftime('%A')  # Weekday name
        if weekday_name in Weekend_days:
            return random.choice(['Monday', 'Tuesday', 'Wednesday', 'Thursday'])
        else:
            return weekday_name

online_shopper_orig['WeekDay'] = online_shopper_orig.apply(get_weekday, axis=1)

```

#Administrative & Administrative_Duration
Scatter plot: This can help visualize the relationship between the number of administrative pages visited and the duration spent on those pages. Each point represents a visit, with the x-axis showing the number of pages and the y-axis showing the duration.



```python

plt.figure(figsize=(8, 6))
plt.scatter(online_shopper_orig['Administrative'], online_shopper_orig['Administrative_Duration'], alpha=0.5)

plt.xlabel('Number of Administrative Pages Visited')
plt.ylabel('Duration Spent on Administrative Pages (Seconds)')
plt.title('Scatter Plot of Administrative Pages vs. Duration')
plt.grid(True)
plt.show()
```

**Positive Correlation**: There appears to be a positive correlation between the number of administrative pages visited and the duration spent on those pages. This means that as the number of pages visited increases, the duration spent on them also tends to increase.


**Data Distribution**: The data points are spread out across the plot, indicating that there is some variability in both the number of pages visited and the duration spent.


**Outliers**: There are a few data points that are further away from the main cluster. These could be potential outliers that warrant further investigation.


**Overall Trend**: The general trend suggests that users who visit more administrative pages tend to spend more time on them.


# Informational & Informational_Duration
Scatter plot: Similar to the Administrative plot, a scatter plot can effectively display the relationship between the number of informational pages visited and the duration spent on them.

```python

plt.figure(figsize=(8, 6))
plt.scatter(online_shopper_orig['Informational'], online_shopper_orig['Informational_Duration'], alpha=0.5)

plt.xlabel('Number of Informational Pages Visited')
plt.ylabel('Duration Spent on Informational Pages (Seconds)')
plt.title('Scatter Plot of Informational Pages vs. Duration')
plt.grid(True)
plt.show()
```

```python
import scipy.stats as stats

# Calculate the Pearson correlation coefficient for the first scatter plot
corr_admin = stats.pearsonr(online_shopper_orig['Administrative'], online_shopper_orig['Administrative_Duration'])[0]

# Calculate the Pearson correlation coefficient for the second scatter plot
corr_info = stats.pearsonr(online_shopper_orig['Informational'], online_shopper_orig['Informational_Duration'])[0]

# Print the results
print(f"Pearson correlation coefficient for Administrative pages: {corr_admin}")
print(f"Pearson correlation coefficient for Informational pages: {corr_info}")
```

A positive correlation coefficient indicates a positive relationship between the variables (as the number of pages visited increases, the duration spent also tends to increase).

The closer the coefficient is to 1, the stronger the positive relationship.

A negative correlation coefficient would indicate a negative relationship (as the number of pages visited increases, the duration spent decreases).

The p-value associated with the correlation coefficient can be used to determine the statistical significance of the observed correlation.

```python
merged_df['Frequency_of_Customer_Logins'] = merged_df['No_Customer_Login'] / merged_df['Session_Activity_Count']

merged_df['Frequency_of_Customer_Logins'] = np.ceil(merged_df['Frequency_of_Customer_Logins'])
```

```python
merged_df['Number_of_Unique_Products_Viewed'] = merged_df.groupby('Session_ID')['Is_Product_Details_viewed'].transform('sum')
```
